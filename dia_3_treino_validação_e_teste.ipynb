```python
#plotando o gráfico para interpretação dos dados faltantes, ou seja, 
#observando a frequência de dados ausentes em cada coluna
def missing_data(df) :
    values_isnull = df.isnull().sum()
    columns = df.columns
    dic = {"colunas":[], "values_isnull":[]}
    for coluna, quant in zip(columns, values_isnull):
        if quant > 0:
            dic['colunas'].append(quant)
            dic['values_isnull'].append(coluna)
    df = pd.DataFrame(dic)
    plt.figure(figsize= (15,5))
    sns.barplot(x= df["values_isnull"], y= df["colunas"], data= df, palette = "tab10")
    plt.xticks(rotation=45);
```

```python
missing_data(spotify_data)
```

8- Pré-processamento dos dados

Objetivo- estabelecer um modelo de machine learning  para prever se uma música será popular ou não.

```python
#removendo dados duplicados

spotify_data= spotify_data.drop_duplicates()
spotify_data.head()
```

```python
#estabelecendo o critério de popularidade

spotify_data[spotify_data["popularity"] >=85]

condition_list= [
    (spotify_data["popularity"] >= 85),
    (spotify_data["popularity"]<85)]

choice_list= [1,0]
spotify_data['most_popular'] = np.select(condition_list, choice_list, default= "Not Specified")
spotify_data['most_popular'] = spotify_data['most_popular'].astype(int)
```

```python
#remover os valores nulos

spotify_data = spotify_data.dropna()
```

```python
spotify_data.columns
```

```python
spotify_data = spotify_data.drop(columns = ["popularity"])
```

```python
#verificando se a classe de popularidade foi criada

spotify_data.info()
```

```python
#limpando os dados, mantendo apenas os dados relevantes para o modelo de machine learning

df_model = spotify_data
cols_to_drop = []
for column in spotify_data:
    if spotify_data[column].dtype == "object":
        cols_to_drop.append(column)
df_model = spotify_data.drop(columns = cols_to_drop)

display(f"extensão dos dados para análise: {df_model.shape}")
```

```python
df_model.head()
```

```python
#remover a coluna 'explicit'

df_model = df_model.drop(columns=["explicit"])
df_model.info()
```

```python
df_model.head()
```

```python
#Transformando as escalas para a normalização dos dados

df_model_nm = (df_model-df_model.min())/(df_model.max()-df_model.min())
```

```python
df_model_nm.head()
```

9- Divisão dos dados e validação cruzada

```python
#importando bibliotecas necessárias
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
```

```python
#separação dos dados
df_model.columns
```

```python
#separação dos dados

x = df_model_nm[['duration_ms', 'danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']]
y = df_model_nm[['most_popular']]

SEED = 280423
np.random.seed(SEED)

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size= 0.25,
                                                        stratify= y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))
```

```python
#taxa de acerto aceitável
from sklearn.dummy import DummyClassifier

dummy_stratified = DummyClassifier(strategy= 'stratified')
dummy_stratified.fit(treino_x, treino_y)
acuracia = dummy_stratified.score(teste_x, teste_y) *100
print("A acuracia do dummy stratified foi de %.2f%%" %acuracia)
```

#para evitar overfitting

```python
#separação dos dados

x = df_treino.drop("most_popular", axis= 1)
y = df_treino.most_popular

#conservando a porcentagem de amostra em cada classe

stratifiedKFold = StratifiedKFold(n_splits=5, shuffle=True, random_state=45)
```

```python
#looping treino/validação

for treino_index, validacao_index in stratifiedKFold.split(x, y):
    x_treino, x_validacao = x.iloc[treino_index], x.iloc[validacao_index]
    y_treino, y_validacao = y.iloc[treino_index], y.iloc[validacao_index]

#visualizando as proporções 

print(f"Dimensões: {x_treino.shape, x_validacao.shape, y_treino.shape, y_validacao.shape}\n")
print(f"Propoção do df_treino para classe=1: {round(len(df_treino[df_treino.most_popular==1])/ df_treino.shape[0],4)}\n")
print(f"Propoção do x_treino para classe=1: {round(len(y_treino[y_treino==1])/x_treino.shape[0],4)}")
print(f"Proporção de x_validacao para classe==1: {round(len(y_validacao[y_validacao==1])/x_validacao.shape[0],4)}")
```
